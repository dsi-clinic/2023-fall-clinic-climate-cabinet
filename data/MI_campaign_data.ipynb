````python
# read in the data depending on the format of the file name., 00 will have the headers while 01 and beyond will not have the headers
# read in the txt files for 2023

def read_txt_data(filepath):
    """
    Reads in the Michigan Campaign Finance Data
    
    Inputs: filepath (str): filepath to the dataset
    
    Returns: df (Pandas DataFrame): Pandas DataFrame from the filepath
    """
    print(f"reading {filepath}")

    if "00" in filepath:
    #if file contains "00" read in the headers 
        df = pd.read_csv(filepath, delimiter="\t", encoding="mac_roman", usecols=["doc_seq_no", "page_no", "contribution_id",
                                                                                            "cont_detail_id", "doc_stmnt_year", "doc_type_desc", 
                                                                                            "com_legal_name", "common_name", "cfr_com_id",	"com_type",	
                                                                                            "can_first_name",	"can_last_name", "contribtype",	"f_name", 
                                                                                            "l_name_or_org", "address",	"city",	"state", "zip",	"occupation", 
                                                                                            "employer",	"received_date", "amount",	"aggregate", "extra_desc"], low_memory=False, on_bad_lines='warn')
    else:
        # files without "00" don't contain headers, set the headers
        df = pd.read_csv(filepath, delimiter="\t", encoding="mac_roman", header=None, names=["doc_seq_no", "page_no", "contribution_id",
                                                                                                "cont_detail_id", "doc_stmnt_year", "doc_type_desc", 
                                                                                                "com_legal_name", "common_name", "cfr_com_id",	"com_type",	
                                                                                                "can_first_name",	"can_last_name", "contribtype",	"f_name", 
                                                                                                "l_name_or_org", "address",	"city",	"state", "zip",	"occupation", 
                                                                                                "employer",	"received_date", "amount",	"aggregate", "extra_desc"], low_memory=False, on_bad_lines='warn')
    print(f"finished reading {filepath}")
    return df
        
    
def concat_years():
    """
    Concatenates the years of election data to create a 1995 to 2023 dataframe
    
    Inputs: None
    
    Returns: merged_df (Pandas DataFrame): Merged dataFrame of MI finance data
    """
    filepath = os.chdir("/Users/necabotheking/Documents/Github/2023-fall-clinic-climate-cabinet/data/Contributions")
    merged_df = pd.DataFrame(columns=range(25))
    
    for file in os.listdir(filepath):
            df = read_txt_data(file)
    
            merged_df.columns = df.columns
            merged_df = pd.concat([merged_df, df])
    return merged_df
````
